```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
conditions = ['Safe Harbor KO', 'RASA2 KO', 'CUL5 KO']
colors = {'Safe Harbor KO': '#a9a9a9', 
          'RASA2 KO': '#800000',
          'CUL5 KO': '#000075'}
```

```python
SH = ['B3', 'B4', 'B5', 'B6']
RASA2 = ['E3', 'E4', 'E5', 'E6']
CUL5 = ['B7', 'B8', 'B9', 'B10']
```

```python
detached_sizes = np.array([123.19, 103.05, 122.08])
detached_sizes_se = np.array([0.23, 0.17, 0.16])
detached_roundness = np.array([0.66174658, 0.645412951, 0.642589604])
detached_roundness_se = np.array([0.00050159, 0.000367118, 0.000346005])
```

```python
attached_sizes = np.array([119.71474122547367, 112.29950659227022, 121.43707234362662])
attached_sizes_se = np.array([0.3249497489731866, 0.3391717216963161, 0.2428184278083937])
attached_roundness = np.array([0.502599034, 0.508483611, 0.471503262])
attached_roundness_se = np.array([0.000974006, 0.000962818, 0.000547423])
```

```python
fig, ax = plt.subplots(1,1,figsize = (7,10))
ax.bar(np.arange(3),
       detached_sizes,
       yerr = 1.96 * detached_sizes_se,
       capsize = 5,
       color = [colors[i] for i in conditions], width = 0.4, label = 'Detached')

ax.bar(0.4 + np.arange(3),
       attached_sizes,
       yerr = 1.96 * attached_sizes_se,
       capsize = 5,
       color = [colors[i] for i in conditions], width = 0.4, alpha = 0.5, label = 'Attached')

ax.set_xticks(np.arange(3) + 0.2, conditions)
ax.set_xlabel('Condition')
plt.ylabel('T-Cell Size (sq. microns)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.legend()
plt.savefig('./figs/t_cell_size_detached_attached.pdf', transparent = True)
plt.show()
```
